```python
columns_companies = list(data_new.columns.values)
corr_mat = pd.DataFrame()

for i in range(data_new.shape[1]):
    row = dict()
    for j in range(data_new.shape[1]):
        
        mid_prof_i = sum(data_new[columns_companies[i]]) / data_new.shape[0]
        mid_prof_j = sum(data_new[columns_companies[j]]) / data_new.shape[0]
        
        numenator = 0
        denominator_i = 0
        denominator_j = 0
        for k in range(data_new.shape[0]):
            numenator += ((data_new[columns_companies[i]][k]- mid_prof_i) * (data_new[columns_companies[j]][k]- mid_prof_j))
            denominator_i += (data_new[columns_companies[i]][k]- mid_prof_i)**2
            denominator_j += (data_new[columns_companies[j]][k]- mid_prof_j)**2
        
                                                            
        correlation = numenator / math.sqrt(denominator_i*denominator_j)
        row[columns_companies[j]] = correlation                                                           
    corr_mat = pd.concat([corr_mat,pd.DataFrame(row, index = [i])], ignore_index= True)

corr_mat                                                                      
```

# Поиск максимальной клики

```python
def bronk2(R, P, X, g):
    if not any((P, X)):
        yield R
    for v in P[:]:
        R_v = R + [v]
        P_v = [v1 for v1 in P if v1 in N(v, g)]
        X_v = [v1 for v1 in X if v1 in N(v, g)]
        for r in bronk2(R_v, P_v, X_v, g):
            yield r
        P.remove(v)
        X.append(v)
def N(v, g):
    return [i for i, n_v in enumerate(g[v]) if n_v]
```

# Поиск максимального независимого множества

```python
#! pip install networkx
import networkx as nx
import networkx.algorithms.approximation as nxaa
import random
 
def maximal_independent_set(G, nodes=None):
 
    if not nodes:
        nodes = set([random.choice(G.nodes())])     # pick a random node
    else:
        nodes = set(nodes)
    if not nodes.issubset(G):
        raise nx.NetworkXUnfeasible("%s is not a subset of the nodes of G" % nodes)
 
    # All neighbors of nodes
    neighbors = set.union(*[set(G.neighbors(v)) for v in nodes])
    if set.intersection(neighbors, nodes):
        raise nx.NetworkXUnfeasible("%s is not an independent set of G" % nodes)
 
    indep_nodes = list(nodes)       # initial
    available_nodes = set(G.nodes()).difference(neighbors.union(nodes)) # available_nodes = all nodes - (nodes + nodes' neighbors)
 
    while available_nodes:
        # pick a random node from the available nodes
        node = random.choice(list(available_nodes))
        indep_nodes.append(node)
 
        available_nodes.difference_update(G.neighbors(node) + [node])   # available_nodes = available_nodes - (node + node's neighbors)
 
    return indep_nodes
```
